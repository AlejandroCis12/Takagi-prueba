```python
import funcion
from funcion import j_newton, T
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def T(x, N=500):
    suma = 0
    for n in range(N):
        tx = 2**n * x
        diente = 2 * min(tx/2 - np.floor(tx/2), np.ceil(tx/2) - tx/2)
        suma += (1/2)**n * diente
    return 0.75 - suma
```

```python
Jz, Jr, Jn, Jc = funcion.j_newton(fun= T, q = 0.5, x0=.3, tol=1e-5, nmax=100)
```

```python
Lx = np.linspace(0, 2, 1000)
Ly = [T(x) for x in Lx]
plt.figure(figsize=(8, 6))
plt.plot(Lx, Ly, linewidth=1.2, label='f(x)')
Jc = np.array(Jc)
plt.plot(Jc[:, 0], Jc[:, 1], '.-', linewidth=1.2, markersize=8, label='Iteraciones')
plt.show()
```
